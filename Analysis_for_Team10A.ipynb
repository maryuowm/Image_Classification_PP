**Import Required Libraries
We'll use pandas for data handling, matplotlib and seaborn for visualization, and gspread to directly load data from Google Sheets.**


```python

# Install required packages
!pip install gspread gspread_dataframe

# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from google.colab import auth
from google.auth import default
import gspread
from gspread_dataframe import get_as_dataframe



```

**Authenticate and Access Google Sheets**
Authenticating Google Sheets access in Colab to read the data directly into a DataFrame.

```python
# Authenticate and authorize
auth.authenticate_user()
creds, _ = default()
client = gspread.authorize(creds)

# Access the sheet by name
spreadsheet = client.open("Copy of Team 10A_Cleaned Dataset v2")
worksheet = spreadsheet.sheet1
data = get_as_dataframe(worksheet)
```

**Data Exploration and Analysis**
With the data loaded, we can now perform some analyses.

**Age Distribution of Learners**
We'll look at the distribution of learner ages to understand the age demographics.

```python
# Check if age column exists and is numeric
if 'age calculation' in data.columns:
    plt.figure(figsize=(10, 6))
    sns.histplot(data['age calculation'], bins=10, kde=True)
    plt.title('Age Distribution of Learners')
    plt.xlabel('Age')
    plt.ylabel('Frequency')
    plt.show()
else:
    print("The 'age calculation' column is missing in the dataset.")
```

```python
from datetime import datetime

# Verify if 'birth_date' column exists and convert it to datetime
if 'Date of Birth' in data.columns:
    data['Date of Birth'] = pd.to_datetime(data['Date of Birth'], errors='coerce')

    # Calculate age
    today = datetime.today()
    data['age'] = data['Date of Birth'].apply(lambda x: today.year - x.year if pd.notnull(x) else None)
else:
    print("The 'Date of Birth' column is missing in the dataset.")
```

**Plotting the Age Distribution**

```python
# Plot the age distribution
plt.figure(figsize=(10, 6))
sns.histplot(data['age'].dropna(), bins=10, kde=True)
plt.title('Age Distribution of Learners')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

**Summary Statistics of Age**

```python
# Check for any null values and then calculate statistics
age_stats = data['age'].dropna().describe()
print("Summary Statistics for Age:")
print(age_stats)
```

**Age Group Distribution**
Categorizing ages into defined groups and plotting them.

```python
# Define age groups
bins = [0, 17, 25, 35, 45, 55, 65, 100]
labels = ['<18', '18-25', '26-35', '36-45', '46-55', '56-65', '66+']
data['age_group'] = pd.cut(data['age'], bins=bins, labels=labels)

# Plot the age group distribution
plt.figure(figsize=(10, 6))
sns.countplot(x='age_group', data=data, palette='coolwarm')
plt.title('Age Group Distribution of Learners')
plt.xlabel('Age Group')
plt.ylabel('Number of Learners')
plt.show()
```

**Age vs. Sign-Up Month**
Analysis on age groups and if a specific agegroup is more likely to sign up in certain months.
 This box plot helps to see if certain months have a higher concentration of particular age groups signing up.

```python
if 'Learner SignUp DateTime' in data.columns:
    data['Learner SignUp DateTime'] = pd.to_datetime(data['Learner SignUp DateTime'], errors='coerce')
    data['sign_up_month'] = data['Learner SignUp DateTime'].dt.month

    # Plot age distribution by sign-up month
    plt.figure(figsize=(12, 8))
    sns.boxplot(x='sign_up_month', y='age', data=data, palette='viridis')
    plt.title('Age Distribution by Sign-Up Month')
    plt.xlabel('Sign-Up Month')
    plt.ylabel('Age')
    plt.show()
else:
    print("The 'Learner SignUp DateTime' column is missing in the dataset.")
```

**Age vs. Gender Distribution**  The violin plot shows the age distribution within each gender, highlighting any patterns or concentrations.

```python
if 'Gender' in data.columns:
    plt.figure(figsize=(12, 8))
    sns.violinplot(x='Gender', y='age', data=data, palette='pastel')
    plt.title('Age Distribution by Gender')
    plt.xlabel('Gender')
    plt.ylabel('Age')
    plt.show()
else:
    print("The 'Gender' column is missing in the dataset.")
```

**Month with the Highest Sign-Ups**
with the Learner SignUp DateTime column, we can parse it to find the month with the highest learner sign-ups.

```python
# Convert sign-up date to datetime and extract month
if 'Learner SignUp DateTime' in data.columns:
    data['Learner SignUp DateTime'] = pd.to_datetime(data['Learner SignUp DateTime'], errors='coerce')
    data['sign_up_month'] = data['Learner SignUp DateTime'].dt.month

    # Plot the number of sign-ups per month
    plt.figure(figsize=(10, 6))
    sns.countplot(x='sign_up_month', data=data, palette='viridis')
    plt.title('Monthly Sign-Ups')
    plt.xlabel('Month')
    plt.ylabel('Number of Sign-Ups')
    plt.show()

    # Display the month with the highest sign-ups
    most_signups_month = data['sign_up_month'].value_counts().idxmax()
    print(f"The month with the highest sign-ups is: {most_signups_month}")
else:
    print("The 'Learner SignUp DateTime' column is missing in the dataset.")
```

**Sign-Ups Over Time**
This analysis helps understand the trend of learner sign-ups over time.

```python
if 'Learner SignUp DateTime' in data.columns:
    data['sign_up_year_month'] = data['Learner SignUp DateTime'].dt.to_period('M')
    monthly_signups = data['sign_up_year_month'].value_counts().sort_index()

    plt.figure(figsize=(12, 6))
    monthly_signups.plot(kind='line', marker='o')
    plt.title('Sign-Ups Over Time')
    plt.xlabel('Year-Month')
    plt.ylabel('Number of Sign-Ups')
    plt.grid(True)
    plt.show()
else:
    print("The 'Learner SignUp DateTime' column is missing in the dataset.")
```

# **Churn analysis and Predictive modeling.**
This report is outlininh the predictive models built, their performance, and the key factors influencing student drop-offs.



```python
# Import libraries
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import LabelEncoder

# Load the data
sheet = client.open_by_url("https://docs.google.com/spreadsheets/d/1OsZwaUiab-9O0AN4i4G9BHfZpT2_xBXAEnAZhtGY_ic/edit?usp=sharing").sheet1

# Get all values from the sheet as a list of lists
data_list = sheet.get_all_values()

# Convert the list of lists to a DataFrame
data = pd.DataFrame(data_list[1:], columns=data_list[0])

# Inspect the data
print("Data Information:")
data.info()
print("\nData Head:")
print(data.head())
```

**Encoding Categorical Variables -** Converting categorical features to numeric.

```python
# Encode categorical features
le = LabelEncoder()


categorical_features = data.select_dtypes(include=['object']).columns.tolist()

# Print categorical_features to check the names
print(f"Categorical features: {categorical_features}")

for col in categorical_features:
    if col in data.columns:  # Checking if column exists before processing
        data[col] = le.fit_transform(data[col])
    else:
        print(f"Column '{col}' not found in the DataFrame. Skipping...")

# Drop rows with missing target
if 'Opportunity Category' in data.columns:
    data = data.dropna(subset=['Opportunity Category'])
else:
    print("Target column 'Opportunity Category' not found. Skipping row drop.")
```
